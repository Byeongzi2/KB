```python
score1 = pd.DataFrame (
    {
        "이름" : ["홍","김"],
        "중간영어" : [70,90],
        "중간수학" : [50,80],
    }
)
score2 = pd.DataFrame (
    {
        "이름" : ["홍","김"],
        "기말영어" : [80,90],
        "기말수학" : [60,70],
    },index=[3,4]
)
```

```python
display(score1)
display(score2)
```

```python
pd.concat([score1,score2]) 
# 세로방향으로 합쳐서 각각 가지고 있지 않은 점수들은 NaN값 처리됨.
```

```python
 pd.concat([score1,score2],axis=1,join='inner') 
# 인덱스 값을 맞춰야 세로로 늘여뜨릴 수 있음.
```

```python
gas1.append(gas2) 
# append : 데이터프레임이 가지고 있는 함수.
# gas1 밑에다가 gas2를 추가시키는 것.
# concat : pandas에 있는 함수.
```

```python
pd.merge(score1,score2) # 이름을 기준으로 병합.
# score2에서 '이름' 칼럼을 '학생이름'으로 바꾸면 에러발생.
```

```python
score1 = pd.DataFrame (
    {
        "이름" : ["홍","김"],
        "중간영어" : [70,90],
        "중간수학" : [50,80],
    }
)
score2 = pd.DataFrame (
    {
        "학생이름" : ["홍","김"],
        "기말영어" : [80,90],
        "기말수학" : [60,70],
    },index=[3,4]
)
```

```python
pd.merge(score1,score2, # on="이름"
        left_on="이름",right_on = "학생이름") 
# 이름이 중복되는 칼럼이 여러개면 오류남. 
# 칼럼이름이 여러개가 같다면 on이라는 속성이용해서 기준 칼럼 지정.
```
