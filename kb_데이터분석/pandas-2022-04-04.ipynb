```python
titanic["age"].fillna(26.9)
# 값이 나왔다는 거는 원본이 바뀐게 아니고 26.9를 집어넣은 복사본을 출력.
```

```python
# 원본에 반영시키는 방법 2가지

# 복사본 만들고 원본에 있는 결측치 값을 복사본에 있는 값으로 대입
titanic["age"] = titanic["age"].fillna(26.9)
# 복사본 안만들고 원본에 직접 대입.
titanic["age"].fillna(26.9, inplace = True)
```

```python
b = titanic["age"].isna()
age_na_idx = titanic[b].index
# titanic[b].index => age에 해당하는 value값들 중 결측치를 가지는 index값 저장
age_na_idx
titanic.loc[age_na_idx,:] # 행 인덱스기 때문에 loc 사용해야 함.
```

```python
titanic.who.len() # who라는 칼럼에 있는 각각의 항목의 len이라는 함수가 적용된 결과를 알고싶다는 의미.
#이렇게 쓰면 who 칼럼 하나는 Series 타입인데, Series에는 len이라는 함수가 없음. So, 오류 발생.
# who라는 칼럼이 가진 len이라는 함수를 가져오세요.라는 의미가 되는 것임. 
# So, 이 오류를 해결하려면 Series에는 str이라는 속성이 있음. So, str이 내부적으로 각 항목에 len을 대신 호출해줌
# 그 결과를 항목의 갯수만큼의 Series로 만들어서 다시 되돌려줌. 
# 원래 의도를 제대로 충족시키려면 아래 셀 코드처럼 해야함. 
```

```python
# who라는 칼럼에 있는 각각의 항목의 len이라는 함수가 적용된 결과를 알고 싶다는 코드
titanic.who.str.len() # 각 문자열 항목에 len을 대신 호출해주고 그 호출 결과를 한줄짜리 Series로 반환해서 보여줌.
```

```python
sp2 = sp.reset_index() # sp를 바꾼게 아니고 sp를 reset한 새로운 data frame이 만들어짐.
sp2
# index는 0,1,2..가 되버리고 sex, pclass는 더 이상 index(=key)값이 아니게 됨.
# 일반 column으로 전락.
```
