```python
url="https://www.kovo.co.kr/game/v-league/11110_schedule_list.asp"
import requests
r = requests.get(url=url)

from bs4 import BeautifulSoup
soup = BeautifulSoup(r.text, "html.parser")

tr_arr = soup.select("table.lst_board.lst_schedule > tbody > tr")
print(len(tr_arr))
for tr in tr_arr:
    dt = tr.select_one(".date").text
    game_info = ""
    right = tr.select_one(".tright") 
    if right : # right에 값이 존재하면
        game_info += right.text
    left = tr.select_one(".tleft") 
    if left: # left에 값이 존재하면
        game_info += left.text
    print(dt, game_info)
```
